### BoT-IoT Dataset

<div style="text-align: justify">
    The <a href="https://research.unsw.edu.au/projects/bot-iot-dataset">BoT-IoT dataset</a> was used in this research to perform the analysis of ensemble methods for evolving data stream classification. The dataset was released in 2018 by the Cyber Range Lab of UNSW Canberra and presents realistic IoT network traffic that incorporates a combination of normal and attack instances.<br><br>
    The testbed used to construct the dataset consists of three components: (i) network platforms, that include normal and attacking virtual machines, an Ubuntu server, Ubuntu mobile, Windows 7, Metasploitable and an Ubuntu Tap machine; (ii) simulated IoT services deployed in the Ubuntu server, that reproduce a weather station, a smart fridge, motion activated lights, a remotely activated garage door and a smart thermostat. To simulate the network behavior of the devices, the authors employed the <a "href=https://nodered.org/">Node-red tool</a>. Also, the five scenarios were connected to the Ubuntu server where a <a href="https://mosquitto.org/">Mosquitto MQTT broker</a> was installed, and to the <a href="https://aws.amazon.com/iot/">AWS IoT hub</a>; and (iii) extracting features and forensics analytics, where the <a href="https://openargus.org/">Argus tool</a> was used to extract data features, and also statistical models and machine learning techniques were employed to distinguish normal and attack instances.<br><br>
    The dataset original files, including pcap, argus and csv formats, are available <a href="https://cloudstor.aarnet.edu.au/plus/s/umT99TnxvbpkkoE">here</a>. The captured pcap files are 69.4 GB in size, with more than 72 million records. The extracted flow traffic, in csv format, is 17.4 GB in size. It includes Service Scanning, OS Fingerprint, DDoS, DoS, Keylogging and Data Theft attacks, with the DDoS and DoS attacks also divided by protocol.
</div>

```python
csv_file = "../processed-data/CSV/botiot-10-eda.csv"
df = pd.read_csv(csv_file)
```

```python
df.info()
```

<div style="text-align: justify">
    Besides the 35 original features, an initial processing of the dataset included 24 new features that represent the source and destination IP addresses. Each of the IP addresses (source and destination) is represented by 12 features. If the protocol of the address is IPv4, the four features that represent the IPv4 are filled with the decimal value of each position of the address and the eight features that represent the IPv6 are filled with (-1). If the protocol of the address is IPv6, the eight features that represent the IPv6 are filled with the conversion from hexadecimal to decimal of the value in each position of the address and the four features that represent the IPv4 are filled with (-1).<br>

Since this process were performed, the original features <b>saddr</b> and <b>daddr</b> can be removed.
</div>

```python
df = df.drop(["saddr", "daddr"], axis=1)
```

<div style="text-align: justify">
     The first approach is to search for features that are not used and thus have only null values, since these features can be removed.
</div>

```python
nullvalues = df.isnull().sum()
print(nullvalues[nullvalues > 0])
```

Since **smac**, **dmac**, **soui**, **doui**, **sco** and **dco** are null for all instances, those features can be removed.

```python
df = df.drop(["smac", "dmac", "soui", "doui", "sco", "dco"], axis=1)
```

The data type of the numeric features **sport** and **dport**, that currently have the type "object", has to be adjusted.

```python
df["sport"] = pd.to_numeric(df["sport"], errors="coerce")
df["dport"] = pd.to_numeric(df["dport"], errors="coerce")
```

Also, the rows where the values of **sport** and **dport** are null can be replaced by (-1).

```python
df = df.fillna(-1)
```

After the removal of the features and the replacement of the remaining nulls, it is possible to observe that there are no more missing values left.

```python
df.isnull().sum().sum()
```

Besides that, as this research is only interested at classify an instance as attack or normal traffic, the features **category** and **subcategory** can be removed.

```python
df = df.drop(["category", "subcategory"], axis=1)
```

<div style="text-align: justify">
    At this point, three categorical features remain: <b>flgs</b>, <b>proto</b> and <b>state</b>. To evaluate if it is possible to perform one-hot encoding, the number of unique values in each of these features is verified.
</div>

```python
print("Columns with Dtype == Object: " +  str(df.select_dtypes('O').columns.to_list()))
print()

print("## Unique Values ##")
print()
print("flgs              " + str(df["flgs"].nunique()))
print("proto             " + str(df["proto"].nunique()))
print("state             " + str(df["state"].nunique()))
```

<div style="text-align: justify">
    Since features <b>flgs</b>, <b>proto</b> and <b>state</b>, which have the type "object", have few unique values, it is possible to convert those into dummies, through one-hot encoding, and then remove the original features.
</div>

```python
df = pd.concat([df, pd.get_dummies(df["flgs"])], axis=1)
df = pd.concat([df, pd.get_dummies(df["proto"])], axis=1)
df = pd.concat([df, pd.get_dummies(df["state"])], axis=1)

df = df.drop(["flgs", "proto", "state"], axis=1)
```

```python
# Removes spaces from the name of the new dummy columns
df.columns = df.columns.str.replace(' ', '')
```

Finally, the feature **attack** is placed at the last position of the dataframe, since it is the label of the data.

```python
df["label"] = df["attack"]
df = df.drop(["attack"], axis=1)
```

To proceed with the EDA, the indexes of the dataframe are reseted, considering that some columns were removed.

```python
# Resets indexes for visualization purposes only
df = df.reset_index(drop=True)
```

Finally, the following features remain.

```python
df.info()
```

<div style="text-align: justify">
    Hence, this EDA will continue from here with two different sets of data. First one is the unbalanced dataset. For the second one, a pipeline that simultaneously applies a random over-sampling technique and a random under-sampling method is used to balance the number of instances.
</div>

```python
unbalanced = df
```

```python
X = df.drop("label", axis=1)
y = df["label"]

over = RandomOverSampler(sampling_strategy=0.4)

under = RandomUnderSampler(sampling_strategy=0.8)

steps = [('o', over), ('u', under)]

pipeline = Pipeline(steps=steps)

X_resampled, y_resampled = pipeline.fit_resample(X, y)

balanced = pd.concat([X_resampled, y_resampled], axis=1)
```

```python
balanced["label"].value_counts()
```

```python
p = sns.countplot(x="label", data=balanced, palette=colors)
p.set(xlabel="Class", ylabel="Count")
sns.despine()
plt.show()
```

Now, the correlation between all features is analysed for both datasets. To do so, the matrix containing the Spearman's rank correlation coefficient is used.

<center>
    <div style="text-align: center;width: 90%;padding: 25px;border-radius: 25px;background-color: #ffecb3;border: 4px solid #e0f7fa">
        The correlation coefficient of Spearman is a statistical measure that can evaluate a monotonic relationship between two features.<br><br>
        <a href="https://mathworld.wolfram.com/MonotonicFunction.html">A monotonic function is a function which is either entirely nonincreasing or nondecreasing.</a><br>
        But not necessarily at a constant rate as it happens in a linear function.<br><br>
        The correlation coefficient between two features is expressed as a number between (-1) and (+1), where it is equal to (-1) for the strongest negative correlation, it is equal to (+1) for the strongest positive correlation and it is equal to 0 if there is no correlation at all.
    </div>
</center>

<div style="text-align: justify">
    Considering this, a threshold <b><i>t = 0,6</i></b> was established for the value of the correlation between any feature and the label. Meaning that, if the absolute value of the correlation between a feature and the label is less than <i>t</i>, this feature can be eliminated/droped, since it possesses a weak relationship with the target.
</div>

```python
threshold = 0.6
```

#### Unbalanced Dataset

```python
corr_u = unbalanced.corr(method="spearman")
corr_u
```

```python
corr_u_label = abs(corr_u["label"])

strong_features_u = corr_u_label[corr_u_label >= threshold]
strong_features_u.sort_values(ascending=False)
```

#### Balanced Dataset

```python
corr_b = balanced.corr(method="spearman")
corr_b
```

```python
corr_b_label = abs(corr_b["label"])

strong_features_b = corr_b_label[corr_b_label >= threshold]
strong_features_b.sort_values(ascending=False)
```

<div style="text-align: justify">
    As this research intends to explore ensemble methods whose base learner is the Hoeffding Tree, the evaluation of decision trees constructed using different combinations of the four remaining balanced dataset's features for the classification of the data between normal and attack are presented next.
```python
X = stime.drop(["label"], axis = 1)
y = stime[["label"]]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)

clf = DecisionTreeClassifier()

clf = clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)

cm = metrics.confusion_matrix(y_test, y_pred)

group_names = ["True Neg", "False Pos", "False Neg", "True Pos"]
group_counts = ["{0:0.0f}".format(value) for value in cm.flatten()]
group_percentages = ["{0:.2%}".format(value) for value in cm.flatten() / np.sum(cm)]
labels = [f"{v1}\n{v2}\n{v3}" for v1, v2, v3 in zip(group_names, group_counts, group_percentages)]
labels = np.asarray(labels).reshape(2, 2)

sns.heatmap(cm, annot=labels, fmt="", cmap=cmap_cm)
plt.show()

accuracy = round((metrics.accuracy_score(y_test, y_pred) * 100), 2)
precision = round((metrics.precision_score(y_test, y_pred) * 100), 2)
recall = round((metrics.recall_score(y_test, y_pred) * 100), 2)

table.add_row(["stime",
               f"{accuracy:.2f}",
               f"{precision:.2f}",
               f"{recall:.2f}"])

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
```
